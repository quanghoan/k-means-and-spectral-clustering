```python
img,mat = ImageAndSegmentations ("271031")
hight = img.shape[0]
width = img.shape[1]
```

```python
def From3DTo2D (img):
    new_img = img.reshape(((img.shape[0]*img.shape[1])), img.shape[2])
    return new_img

def From2DTo3D(image,hight,width):
    return image.reshape((hight,width))
    
```

```python
def ClusterIndicesNumpy(clustNum, labels_array): 
    return np.where(labels_array == clustNum)[0]
```

# Segmentation using K-means 

```python
Kmean_img = From3DTo2D (img)
Kmean_img.shape
```

```python
for i in range (3,12,2):
    kmeans = KMeans(n_clusters = i).fit(Kmean_img)
    label = kmeans.labels_
    image =  From2DTo3D(label,hight,width)
    print ("Number Of Clusters = ",i)
    plt.imshow(image)
    plt.show()
```

```python
def ClusteredLabeled (k,img):
    kmeans = KMeans(n_clusters = k).fit(img)
    label = kmeans.labels_
    cluster = []
    clusters = {}
    for i in range(kmeans.n_clusters):
        index = ClusterIndicesNumpy(i,label)
        cluster = []
        for k in index:
            cluster.append(img[k])
        clusters.update([(i,cluster)])
    return clusters
```

# F1 Score Evaluation

```python
f1score = 0
counter = 0
for i in range (3,12,2):
    counter = 0
    f1score = 0
    kmeans = KMeans(n_clusters = i).fit(Kmean_img)
    label = kmeans.labels_
    for k in range (10):
        try:
            x = mat['groundTruth'][0][k][0][0][0]
            x = x.reshape(hight*width)
            f1score = f1score + f1_score(x,label, average='micro')
            counter = counter +1
            #print(f1score)
        except IndexError:
                pass
        continue
    #print(counter)
    print("For K = ",i,"The F1 score is :",f1score/counter)
```

# Condetional Entropy (Implemented From Scratches)

```python
def Condetional_Entropy(label):
    numbers = []
    lenght = []  
    for i in range (10):
        try:
            x = mat['groundTruth'][0][i][0][0][0]
            x = x.reshape(hight*width)
            for k in range(kmeans.n_clusters):
                index = ClusterIndicesNumpy(k, label) #getting the indcies of cluster 0 and 1 and 2 in case number of clusters is 3
                #print(label)
                #print(k,len(index))
                lenght.append(len(index))
                counter = 0
                #print("for cluster ",k)
                for i in index:
                    if (x[i]==k+1):
                        counter = counter + 1
                numbers.append(counter)
        except IndexError:
               pass
        continue
    return numbers,lenght
```

```python
for j in range (3,12,2):
    kmeans = KMeans(n_clusters = j).fit(Kmean_img)
    label = kmeans.labels_
    numbers,lenght = Condetional_Entropy(label)
    CE = 0
    for k in range (kmeans.n_clusters):
        zeros = []
        for i in range(k,len(numbers),kmeans.n_clusters):
            zeros.append(numbers[i])
        maximum = max(zeros)
        #print (k,maximum)
        #print (lenght[i])
        if (maximum!=0 and lenght[i]!=0):
            CE = CE - (maximum/lenght[i])*log10((maximum/lenght[i]))
    print ("The Condetional Entropy When K = ",j,"Will Equal To :",CE)
```

# Implementation of K-Means clustring from scratches

```python
def IntilizingCentroids(numberofclusters,image):
    centeroids = {}
    for i in range (numberofclusters):
        centeroids[i]= image[randrange(200)]
    return  centeroids
```

```python
def K_MEANS(number_of_clusters,image):
    Centeroids = IntilizingCentroids(number_of_clusters,image)
    for i in range(300):
        classification = {}
        distances = []
        for i in range (number_of_clusters):
            classification[i] = []
        for RGB in image :
            for i in Centeroids:
                distances.append(np.linalg.norm((RGB-Centeroids[i])))
             #   print(distances)
            index = distances.index(min(distances))
            #print(index)
            classification[index].append(RGB)
            distances = []
        Origenal_Centeroids = dict(Centeroids)
        for i in range (number_of_clusters):
            Centeroids[i] = np.mean(classification[i],axis = 0)
           # print(Centeroids[i])
        for i in range(number_of_clusters):
            if(np.sum((Centeroids[i]-Origenal_Centeroids[i])/Origenal_Centeroids[i]*100.0)<0.0001):
                break
            if (np.array_equal(Origenal_Centeroids,Centeroids)):
                break
    return classification
```

# Spectral Clustering Segmentation

```python
def ResizeImage(image,hight,width):
    os.chdir(r"C:\Users\Nayera Hatem\BSR\BSDS500\data\images\train")
    im = Image.open(image+".jpg")
    y = int(hight/4)
    x = int(width/4)
    im =  im.resize((x,y))
    image = np.asarray(im)
    return image,y,x   
```

```python
image,h,w = ResizeImage("2092",hight,width)
plt.imshow(image)
image = From3DTo2D(image)
```

```python
clustering = SpectralClustering(n_clusters=5,affinity='nearest_neighbors',assign_labels="discretize",).fit(image)
label = clustering.labels_
image =  From2DTo3D(label,h,w)
print ("Number Of Clusters = ",5)
plt.imshow(image)
plt.show()
```
